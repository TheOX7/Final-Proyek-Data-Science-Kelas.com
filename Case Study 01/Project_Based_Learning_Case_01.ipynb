```python
#Import Libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Create DataFrame
df = pd.DataFrame({
    'TransactionAmount' : [100, 150, 50, 100, 130, 120, 100, 85, 70, 150,
                           150, 120, 50, 100, 100, 140, 90, 150, 50, 90,
                           120, 100, 110, 75, 65]
})

df.head()
```

```python
#Central Tendency
import numpy as np
print(f"Mean : {np.mean(df['TransactionAmount'])}")
print(f"Median : {np.median(df['TransactionAmount'])}")
```

```python
# Tulis analisismu dari Central Tendency di sini
# Nilai mean > median, mengartikan bahwa terdapat nilai-nilai yang lebih tinggi dibandingkan dengan value lainnya.
# Namun sebenarnya nilai mean dan median ini tidak terlalu jauh perbedaannya. Oleh karena itu menjukkan bahwa data terdistribusi cukup seimbang
```

```python
#Variability
print(f"Variance : {df['TransactionAmount'].var()}")
print(f"Std. Deviation : {df['TransactionAmount'].std()}")
print(f"Max : {df['TransactionAmount'].max()} | Min : {df['TransactionAmount'].min()} ")
print(f"Q1 : {df['TransactionAmount'].quantile(0.25)} | Q2 : {df['TransactionAmount'].quantile(0.5)} | | Q3 : {df['TransactionAmount'].quantile(0.75)}")
```

```python
#Tulis analisismu dari Variability di sini
```

```python
#Define H0 and H1
#H0: Mean transaction amount is less or equals than 102.6
#H1: Mean transaction amount is greater than 102.6
```

```python
#Define alpha = 0.05
alpha = 0.05
```

```python
#Calculate t statistics and P-value
#Perform one sample t-test
mean_value = np.mean(df['TransactionAmount'])
feature = df['TransactionAmount']

t_statistic, p_value = stats.ttest_1samp(feature, mean_value)
```

```python
#statistic= , pvalue=
statistic_result = f"statistic = {t_statistic}, pvalue = {p_value}"
statistic_result
```

```python
#Conclude with alpha and p-value
if p_value < alpha:
    conclusion_alpha_p = "P-Value < alpha. Reject the null hypothesis."
else:
    conclusion_alpha_p = "P-Value >= alpha. Fail to reject the null hypothesis."
    
#Conclusion: 
print(conclusion_alpha_p)
```

```python
#Conclude with t statistics and critical region
#Find the critical region
critical_region = stats.t.ppf(1 - alpha, df=len(feature)-1)
print(critical_region)
```

```python
#H1: ...
h1 = f"H1: Mean transaction amount is greater than {mean_value}"
print(h1)
```

```python
#Critical region result
if t_statistic > critical_region:
    conclusion_critical_region = f"T-Statistic > Critical Region ({critical_region:.4f}). Reject the null hypothesis."
else:
    conclusion_critical_region = f"T-Statistic <= Critical Region ({critical_region:.4f}). Fail to reject the null hypothesis."
    
#Conclusion: 
print(conclusion_critical_region)
```
